### fillna function for replace missing values 


```python
def replacer(df):
    cat = list(df.columns[df.dtypes=='object'])
    for i in df.columns:
        if i in cat:
            md= df[i].mode()[0]
            df[i] = df[i].fillna(md)
        else:
            mn=df[i].mean()
            df[i] = df[i].fillna(mn)
    print('Missing values replaced in DataFrame')
        

```

```python
from sklearn.metrics import median_absolute_error, mean_squared_error, \
                            mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
def evaluate_model(model, x, y):
    # Get the prediction 
    pred = model.predict (x)
    # Calculate all the metrics 
    mse = mean_squared_error (y, pred)
    rmse = mse**(1/2)
    mae = median_absolute_error (y, pred)
    mape = mean_absolute_percentage_error (y, pred)
    r2 = r2_score (y, pred)
    # Print above results
    print(f'Mean squared error : {mse:.2f}')
    print (f"Root mean squared error : {rmse:.2f}")
    print (f"Mean absolute error : {mae:2f}")
    print (f"Meand absolute error percentage {mape:.4f}")
    print (f"R 2 score :{r2:.4f}") 
```

```python
evaluate_model (model, xtrain, ytrain)
```

```python
evaluate_model (model, xtest, ytest)
```

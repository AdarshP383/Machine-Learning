### fillna function for replace missing values 


```python
def replacer(df):
    cat = list(df.columns[df.dtypes=='object'])
    for i in df.columns:
        if i in cat:
            md= df[i].mode()[0]
            df[i] = df[i].fillna(md)
        else:
            mn=df[i].mean()
            df[i] = df[i].fillna(mn)
    print('Missing values replaced in DataFrame')
        

```

```python
replacer(df)
```
